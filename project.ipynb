```python
final_train_set = train_set_numeric.drop("SalePrice",axis=1)
final_train_set_lables = train_set_numeric["SalePrice"].copy()
```

# Creating Imputer object 

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
imputer.fit(final_train_set)
x = imputer.transform(final_train_set)
```

```python
final_data = pd.DataFrame(x,columns=final_train_set.columns)
final_data.describe()
```

# Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy = 'median'))
    ,('std_scaler',StandardScaler())  
])
```

```python
final_data_to_be_pred = my_pipeline.fit_transform(final_train_set)
```

# Create Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor

model_1 = LogisticRegression()
model_2 = DecisionTreeRegressor()

model_1.fit(final_data_to_be_pred,final_train_set_lables)
```

```python
model_2.fit(final_data_to_be_pred,final_train_set_lables)
```

# Testing model on small scale

```python
some_data = final_train_set.iloc[0:5]
some_lables = final_train_set_lables.iloc[0:5]

some_prepared_data = my_pipeline.transform(some_data)

model_1.predict(some_prepared_data)
```

```python
model_2.predict(some_prepared_data)
```

```python
# from sklearn.model_selection import cross_val_score
# final_prepared_data = model_1.predict(final_data_to_be_pred)
# scores = cross_val_score(model_1,final_prepared_data,final_train_set_lables,scoring="neg_mean_squared_error",cv=10)
# rmsc_scores = np.sqrt(-scores)

from sklearn.metrics import mean_squared_error
final_prepared_data = model_1.predict(final_data_to_be_pred)
lin_mse = mean_squared_error(final_train_set_lables,final_prepared_data)
lin_rmsc = np.sqrt(lin_mse)

final_prepared_data = model_2.predict(final_data_to_be_pred)
tree_mse = mean_squared_error(final_train_set_lables,final_prepared_data)
tree_rmsc = np.sqrt(tree_mse)
```
